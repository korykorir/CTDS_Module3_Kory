```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
#create undirected graph
G = nx.Graph()
```

```python
G.add_nodes_from([1,2,3,4])
G.add_edges_from([(1,2),(2,3),(3,4),(4,1)])
```

```python
nx.draw(G, with_labels=True)
plt.show()
```

```python
#create a directed graph
G = nx.DiGraph()
```

```python
G.add_edges_from([(1,2),(2,3),(3,4),(4,1)])
```
